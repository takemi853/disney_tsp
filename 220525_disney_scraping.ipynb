```python
import math
import random
import re

from jmetal.core.problem import PermutationProblem
from jmetal.core.solution import PermutationSolution

"""
.. module:: TSP
   :platform: Unix, Windows
   :synopsis: Single Objective Traveling Salesman problem

.. moduleauthor:: Antonio Benítez-Hidalgo <antonio.b@uma.es>
"""


class TSP(PermutationProblem):
    """Class representing TSP Problem."""

    def __init__(self, instance, rm_atractions):
        super(TSP, self).__init__()

        # 回らない アトラクションを設定
        for rm_atraction in rm_atractions:
          instance = instance.drop(instance[instance[rm_atraction]==0].index)
          instance = instance.drop(rm_atraction,axis = 1)

        distance_matrix = np.array(instance)
        number_of_cities = len(instance)
        # distance_matrix, number_of_cities = self.__read_from_file(instance)

        self.distance_matrix = distance_matrix

        self.obj_directions = [self.MINIMIZE]
        self.number_of_variables = number_of_cities
        self.number_of_objectives = 1
        self.number_of_constraints = 0

        self.round_columns =  instance.columns
        self.round_columns = [i for i in self.round_columns if i not in rm_atractions]

    def evaluate(self, solution: PermutationSolution) -> PermutationSolution:
        fitness = 0

        # 入口(ワールドバザール)から 1つ目のアトラクションまで
        fitness += self.distance_matrix[-1][solution.variables[0]]

        for i in range(self.number_of_variables - 1):
            # print(f'{i} : {fitness}')
            x = solution.variables[i]
            y = solution.variables[i + 1]

            # 営業時間 (9:00~21:00 の 720分)
            if fitness < 720:
              fitness += wait_df[self.round_columns[x]][fitness//15]
            else: #回れなかったときのペナルティ
              fitness += 10000

            fitness += self.distance_matrix[x][y]

        # 最後のアトラクションから 出口(ワールドバザール) まで
        # 最後のアトラクションの待ち時間も計算
        if fitness < 720:
          fitness += wait_df[self.round_columns[y]][fitness//15]
        else: #回れなかったときのペナルティ
          fitness += 10000
        fitness += self.distance_matrix[y][-1]

        solution.objectives[0] = fitness

        return solution

    def create_solution(self) -> PermutationSolution:
        new_solution = PermutationSolution(
            number_of_variables=self.number_of_variables, number_of_objectives=self.number_of_objectives
        )
        new_solution.variables = random.sample(range(self.number_of_variables), k=self.number_of_variables)

        return new_solution

    @property
    def number_of_cities(self):
        return self.number_of_variables

    def get_name(self):
        return "Symmetric TSP"

    # 訪問予定の アトラクション名を取得
    def get_round_attraction_names(self):
        return self.round_columns
```

```python
from jmetal.algorithm.singleobjective.evolution_strategy import EvolutionStrategy
from jmetal.operator.mutation import PermutationSwapMutation
# from jmetal.problem.singleobjective.tsp import TSP
from jmetal.util.termination_criterion import StoppingByTime

if __name__ == "__main__":
    rm_atractions = []

    # park = 'TDL'
    # park = 'TDS'
    # problem = TSP(instance="resources/TSP_instances/kroA100.tsp")
    if park == 'TDL' :
      distance_matrix = pd.read_csv('.//disneyland_移動時間.csv')
      rm_atractions = ['スイスファミリー・ツリーハウス']
    if park == 'TDS' :
      distance_matrix = pd.read_csv('./disneysea_移動時間.csv')
      rm_atractions = ['レールウェイ(フロント)', 'スチーマーライン(ポート)','スチーマーライン(ハーバー)', 'スチーマーライン(ロスト)', 'スチーマーライン(フロント)']
      rm_atractions += ['ジャスミンのフライングカーペット', '海底２万マイル', 'マジックランプシアター'] # 休止アトラクション

    distance_matrix = distance_matrix * 1.5 # 歩く時間 の 設定 (倍率・重み)

    problem = TSP(distance_matrix, rm_atractions)

    print(f"Solving TSP problem with {problem.number_of_cities} cities.")

    algorithm = EvolutionStrategy(
        problem=problem,
        mu=1,
        lambda_=10,
        mutation=PermutationSwapMutation(1.0 / problem.number_of_cities),
        elitist=False,
        termination_criterion=StoppingByTime(60* 1),
    )

    algorithm.run()
    result = algorithm.get_result()

    print(f"Algorithm: {algorithm.get_name()}")
    print(f"Solution: {result.variables}")
    print(f"The shortest path length: {result.objectives[0]}")
    print(f"Computing time: {algorithm.total_computing_time}")
    print(f"Number of evaluations: {algorithm.evaluations}")


    ## 結果 (巡回路) 表示
    fitness = 0
    for rm_atraction in rm_atractions:
      distance_matrix = distance_matrix.drop(distance_matrix[distance_matrix[rm_atraction]==0].index)
      distance_matrix = distance_matrix.drop(rm_atraction,axis = 1)
    columns =  distance_matrix.columns
    columns = [i for i in columns if i not in rm_atractions]

    distance_matrix = np.array(distance_matrix)

    # 入口(ワールドバザール)から 1つ目のアトラクションまで
    fitness += distance_matrix[-1][result.variables[0]]

    for i in range(len(columns) - 1):
        # print(f'{i} : {fitness}')
        x = result.variables[i]
        y = result.variables[i + 1]
        print(f'{i} | {int(9+fitness//60)} : {int(fitness - fitness//60 *60) :02}, {columns[x]}')
        # 営業時間 (9:00~21:00 の 720分)
        if fitness < 720:
          fitness += wait_df[columns[x]][fitness//15]
        else: #回れなかったときのペナルティ
          fitness += 10000

        fitness += distance_matrix[x][y]

    # 最後のアトラクションから 出口(ワールドバザール) まで
    # 最後のアトラクションの待ち時間も計算
    print(f'{i} | {int(9+fitness//60)} : {int(fitness - fitness//60 *60) :02}, {columns[y]}')
    if fitness < 720:
      fitness += wait_df[columns[y]][fitness//15]
    else: #回れなかったときのペナルティ
      fitness += 10000
    fitness += distance_matrix[y][-1]

    print(fitness)
```
